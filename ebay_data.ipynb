```python
import pandas as pd
from pandas import DataFrame, read_csv
import itertools
 
import pprint
```

```python
orders = r'ebayorders.csv'
df = pd.read_csv(orders)
```

```python
df['items_in_order'] = df.groupby('sales_record_number')['sales_record_number'].transform('count')

df['item_occurance_count'] = df.groupby('custom_label')['custom_label'].transform('count')
```

```python
multi_orders = df.loc[df['items_in_order'] > 1]
multi_orders = multi_orders.loc[multi_orders['item_occurance_count'] > 1]
```

```python
import collections
```

```python
c = collections.Counter()
for x, y in multi_orders.groupby('sales_record_number'):
    
    items = list(y['custom_label'])
    
    item_combos = []
    
    item_count = len(items)
    
    for i in range(2,item_count+1):
        item_combos = item_combos + list(itertools.combinations(items,i))
    
    c.update(item_combos)
```
