```python
movie_data = pickle.load(open("mojo_scrape.p", "rb" ))
movies_df = pd.DataFrame(movie_data)
```

We will grab movies made in the last ~25 years, and will only look at movies released before Nov 1 2014 so that the box office totals are reasonably accurate.  We need to grab them in two batches since the budgets are formatted differently for small and large budget movies.

We'll also run a few functions to clean, organize, and pre-process the data for modeling. 

```python
movies_df = get_movies_with_budget(movies_df)
movies_df = deflate_dollar_values(movies_df)
movies_df = separate_players(movies_df)
movies_df, model_genres = add_genres_columns(movies_df)
movies_df, model_actors, model_directors, model_writers = dummify_players(movies_df,5)
```

```python
model_ratings =  ['rating_G','rating_PG','rating_PG-13','rating_R']
dummy_ratings = pd.get_dummies(movies_df['rating'], prefix='rating')
movies_df = movies_df.join(dummy_ratings[model_ratings])

model_seasons =  ['season_summer','season_holiday']
dummy_seasons = pd.get_dummies(movies_df['season'], prefix='season')
movies_df = movies_df.join(dummy_seasons[model_seasons])
```

```python
all_models = {}

#initial model (cols to ALWAYS keep)
x_cols_model = ['log_budget','log_widest_release','ones'] + model_seasons + model_ratings
x_cols_test = model_genres

# the cols we are adding are the cols to test and potentially eliminate
x_cols = x_cols_model + x_cols_test

y = movies_df['log_total_gross']
X = movies_df[x_cols]

linmodel = sm.OLS(y, X).fit()
all_models['initial'] = [linmodel,x_cols]

#now we have our baseline model, let's iterate
num = math.ceil(math.log(len(x_cols)))
p_val_thresholds = np.linspace(.5,.05,num)

for p_val in p_val_thresholds:
    
    cols_to_keep = []
    
    for coef in linmodel.pvalues.iteritems():
        if coef[0] not in x_cols_model:
            if coef[1] < p_val:
                cols_to_keep.append(coef[0])
    
    
    x_cols = x_cols_model + cols_to_keep
    y = movies_df['log_total_gross']
    X = movies_df[x_cols]

    linmodel = sm.OLS(y, X).fit()
    all_models[p_val] = [linmodel,x_cols]

model_genres = cols_to_keep
linmodel.summary()
```

Appendix (code used for plotly formatting/other random, non-function-worthy things)

```python
regression_results = pd.DataFrame(zip(linmodel.fittedvalues, linmodel.resid, linmodel.resid/linmodel.fittedvalues),
                                  columns = ['fittedval','resid','resid/fittedval'])
movies_df = movies_df.join(regression_results)
```

```python
coefs = pd.DataFrame(linmodel.params)

players = model_actors + model_directors + model_writers

player_coefs = coefs.ix[players].dropna()
player_coefs['extreme_vals'] = ''
player_coefs["exp^coeff"] = player_coefs[0].map(lambda x: math.exp(x))
player_coefs['player'] = player_coefs.index.map(
                            lambda x: 
                            x.split('_')[1] +" ("+ x.split('_')[0] +")"
                        )

player_coefs['player'] = player_coefs.index.map(
                            lambda x: 
                            x.split('_')[1] +" ("+ x.split('_')[0] +")"
                        )
player_coefs[['player','exp^coeff']].sort('exp^coeff')


for player in player_coefs['player'].index:

    player_df = movies_df.loc[movies_df[player] == 1]

    coef = player_coefs.ix[player]['exp^coeff']
    if coef > 1:
        qualifier = 'Best Performing Films: '
        
        extremes = player_df[['resid','resid/fittedval','movie_title','budget_int','total_gross']].loc[player_df['resid'] > 0]. \
                        sort('resid/fittedval', ascending=False)[['movie_title']][:3].values.tolist()
    else:
        qualifier = 'Worst Performing Films: '
        
        extremes = player_df[['resid','resid/fittedval','movie_title','budget_int','total_gross']].loc[player_df['resid'] < 0]. \
                        sort('resid/fittedval', ascending=True)[['movie_title']][:3].values.tolist()
        
    extreme_movies = [m[0] for m in extremes]
    string = qualifier + ' | '.join(extreme_movies)
    
    player_coefs['extreme_vals'] = np.where(player_coefs.index == player, string, player_coefs['extreme_vals'])
    
```

```python
player_coefs.sort('exp^coeff')[['exp^coeff','player','extreme_vals']].to_csv('movies_info.csv')
```
